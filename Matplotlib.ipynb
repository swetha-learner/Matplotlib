```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,5,11)
print(x)
y=x**2
print(y)
```

```python
x= np.linspace(0,5,11)
y=x**2

plt.figure()
plt.plot(x,y,color="#417ab0")
plt.xlabel('x')
plt.ylabel('y')
plt.title('sqaure of x')
plt.grid()
plt.axis([-0.5,5.5,-3,28])
plt.show()
```

```python
x= np.linspace(0,5,11)
y=x**2

plt.figure(figsize=[5,5],dpi=120,facecolor='yellow')
plt.plot(x,y,color="#417ab0")
plt.xlabel('x')
plt.ylabel('y')
plt.title('sqaure of x')
plt.grid()
plt.axis([-0.5,5.5,-3,28])
plt.show()
```

```python
x= np.linspace(0,5,11)
y=x**2

plt.figure(dpi=120,facecolor='yellow',linewidth=2,edgecolor='black',frameon=True)
plt.plot(x,y,color="blue",alpha=0.9,linestyle='--',linewidth=2,marker='o',markersize=10)
plt.xlabel('x',fontsize=15)
plt.ylabel('y',fontsize=15)
plt.title('sqaure of x',fontsize=15)
plt.grid(color='r',alpha=0.4)
plt.axis([-0.5,5.5,-3,28])
plt.show()
```

```python
x= np.linspace(0,5,11)
y=x**2

plt.figure()
plt.subplot(1,2,1) #row,col,id
plt.plot(x,y,'r--o') #fmt formating
plt.subplot(1,2,2)
plt.plot(y,x,'g:*')
plt.show()
```

```python
x= np.linspace(0,5,11)
y=x**2
y1=x**3

plt.plot(x,y,color='crimson', label='pwer of 2')
plt.plot(x,y1,color='blue' ,label='power of 3')
plt.legend()
```

# multiple plots

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(0,5,11)
y=x**2
y1=x**3

plt.scatter(x,y,color='crimson', label='pwer of 2')
plt.plot(x,y1,color='blue' ,label='power of 3')
plt.legend()
```

```python
x=['a','b','c']
y=[5,4,2]

z=plt.bar(x,y,color=['r','g','b'],linewidth=2,edgecolor='k',tick_label=["Item"+color for color in x])
print(z)
```

```python
x= np.linspace(0,5,11)
y=x**2

fig=plt.figure()
axes=(fig.add_axes([0.2,0.4,0.6,0.6]))
axes.plot(x,y)
plt.show()
```
